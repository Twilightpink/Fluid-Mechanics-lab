### Introduction :
Bernoulli’s theorem states that when there is a continuous connection between particles of flowing mass of liquid, the total energy at any section of flow will remain the same provided there is no reduction or addition of energy at any point.

### Theory :
Bernoulli’s equation is an energy equation and is based on the law of conservation of energy.  This equation states that at two sections of flow field the total energy remains the same, provided that there is no loss or gain of energy between the two sections.  This equation is valid only for steady flow.  This equation is expressed as:
$$\mathbf{E}=\frac{\mathbf{p}_{1}}{\rho g}+\frac{\mathbf{v}_{1}^{2}}{2 g}+\mathbf{Z}_{1}=\frac{\mathbf{p}_{2}}{\rho g}+\frac{\mathbf{v}_{2}^{2}}{2 g}+\mathbf{Z}_{2}$$
Where:

E           =	Total Energy

$p_{1}$		=	Pressure of fluid at point 1

$p_{2}$		=	Pressure of fluid at point 2

$V_{1}$		=	Velocity of fluid at point 1

$V_{2}$		=	Velocity of fluid at point 2

$\frac{p_{1}}{\rho g}$ =	Pressure energy per unit weight of fluid or pressure head at point. 

$\frac{p_{2}}{\rho g}$ =	Pressure energy per unit weight of fluid or pressure head at point. 

$\frac{V_{1}^{2}}{\rho g}$ =	Kinetic energy per unit weight or kinetic head at point 1

$\frac{V_{2}^{2}}{\rho g}$ 	=	Kinetic energy per unit weight or kinetic head at point 2

Z1		=	Potential energy per unit weight of potential head at point 1

Z2		=	Potential energy per unit weight of potential head at point 2

$\rho$		=	Density of fluid.

g		=	Acceleration due to Gravity

Pressure head is measured by piezometer tube, so all the points are at same horizontal level, so potential head is constant for all points and can be taken as zero. So Bernoulli’s Equation can be written as,

$$\textbf{E} = \textbf{h} + \frac{\textbf{V}^{2}}{2\textbf{g}}$$

where $$ \textbf{h} = \frac{p}{\rho g}$$

### Calculations:

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
Q = np.array([820,730,560,600,660,480])                #lph
pz1 = np.array([26.4,20.4,35,31.8,26,35.6])
pz2 = np.array([26.2,20.2,34.8,31.6,26,35.4])
pz3 = np.array([24.2,18.4,34.2,30.2,24.4,34.8])
pz4 = np.array([20.8,15.6,32.2,28.2,21.8,32.8])
pz5 = np.array([13.4,9.4,28,23.8,16.6,29])
pz6 = np.array([18.2,13.2,30,25.8,19.8,30.6])
pz7 = np.array([21.4,16,32.2,28.4,22,32.8])
pz8 = np.array([22.8,17,33,29.4,23.2,33.6])
pz9 = np.array([23.2,17.4,33,29.2,23.4,33.8])
g = 9.81
ac = np.array([6.1575,4.7143,3.463,2.405,1.539,2.405,3.463,4.7143,6.1575])*1e-4
```

```python
obs = pd.DataFrame({'LPH':Q,'pz1':pz1,'pz2':pz2,'pz3':pz3,'pz4':pz4,'pz5':pz5,'pz6':pz6,'pz7':pz7,'pz8':pz8,'pz8':pz8,
                   'pz9':pz9})
print(obs)
```

```python
pzhead = obs.to_numpy().T[1:].T*1e-2                                  #Pressure Head
print(pzhead)
vhead = np.array([(elem*1e-3/(ac*3600))**2/(2*9.8) for elem in Q])    #V
thead = vhead+pzhead                                                  #Total energy
print("The total energy at different points is {} m".format(thead))
```

```python
plt.figure()
for elem,f in zip(vhead,Q):
    plt.plot(elem,label=f)
plt.grid()
plt.legend()
plt.xlabel('Piezometer tube number')
plt.ylabel('Velocity head (m)')
```

```python
plt.figure()
for elem,f in zip(pzhead,Q):
    plt.plot(elem,label=f)
plt.grid()
plt.legend()
plt.xlabel('Piezometer tube number')
plt.ylabel('Pressure head (m)')
```

```python
plt.figure()
for elem,f in zip(thead,Q):
    plt.plot(elem,label=f)
plt.grid()
plt.legend()
plt.xlabel('Piezometer tube number')
plt.ylabel('Total head (m)')
```

### Inferences:
1. From the experiment it can be concluded that with decrease in area of flow there is an increase in velocity and decrease in the flow pressure of the fluid. 
2. The velocity peaks at the converging part of the nozzle and then again decreases.
3. Therefore, it can be concluded that the Bernoulli’s equation is valid when applied to steady flow of water in tapered duct and absolute velocity values increases along the same channel and then decreases.
